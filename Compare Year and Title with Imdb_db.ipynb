```python
import pandas as pd
```

```python
combined_df = pd.read_csv('./Year_Released Title and Metacritic Rating.csv')
imdb_db = pd.read_csv('./imdb_db.csv')
```

```python
imdb_db = imdb_db[['primaryTitle','averageRating','startYear']]
imdb_db = pd.DataFrame(imdb_db)
imdb_db.columns = ['Title','Imdb Rating','Year']
titles = titles[['Title','Year']]
```

```python
combined_df = combined_df.loc[combined_df['Year'] != 'TBA']
```

```python
combined = pd.merge(combined_df,imdb_db,how='inner',on = 'Title')
```

```python
combined['Year_x'] = combined['Year_x'].astype(int)
combined['Year_y'] = combined['Year_y'].astype(int)
```

```python
combined = combined[combined['Year_x'] == combined['Year_y']]
```
